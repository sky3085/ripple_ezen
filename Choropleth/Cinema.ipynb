```python
import pandas as pd
```

```python
df = pd.read_csv("data/Seoul_Cinema_2022.csv", encoding="euc-kr")
df = df["영업상태명 전화번호 지번주소 도로명주소 사업장명 좌표정보(X) 좌표정보(Y)".split()]
print(len(df))
df.head()
```

```python
# 취소/말소/만료/정지/중지, 폐업 제거
df.drop(df[df['영업상태명'] == "폐업"].index, inplace=True)
df.drop(df[df['영업상태명'] == "취소/말소/만료/정지/중지"].index, inplace=True)
df.drop(['영업상태명'], axis=1, inplace=True)
print(len(df))
df.head(3)
```

```python
# 같은 지점 제거
df.drop_duplicates(['좌표정보(X)', '좌표정보(Y)'], inplace=True)
print(len(df))
df.head(10)
```

```python
df.reset_index(inplace=True, drop=True)
df.head(3)
```

```python
import re
```

```python
df_name = df['사업장명']
df_name = df_name.str.replace("(", " ")
df_name = df_name.str.replace(")", "")
df_name = df_name.str.replace("제", " ")
df_name = df_name.str.split()
for idx, name in enumerate(df_name):
    if("관" in name[-1]):
        name = name[:-1]
    name = " ".join(name)
    df_name[idx] = name
print(df_name)
```

```python
df['사업장명'] = df_name
df.head(10)
# X :127.043930, Y :37.6651626
# 203799.865284	462533.430211
```

```python
df['전화번호'].fillna("-", inplace=True)
```

```python
df.to_csv("data/Seoul_Cinema_2022_fix", index=False)
```

```python
# 중부원점 -> 경위도로 변환
from pyproj import Proj, transform

# Projection 정의
# 중부원점(Bessel): 서울 등 중부지역 EPSG:2097
proj_1 = Proj(init='epsg:2097')

# WGS84 경위도: GPS가 사용하는 좌표계 EPSG:4326
proj_2 = Proj(init='epsg:4326')

lon = []
lat = []
for idx in df.index:
    x = df['좌표정보(X)'][idx]
    y = df['좌표정보(Y)'][idx]
    converted = transform(proj_1, proj_2, x, y)
    lon.append(converted[0])
    lat.append(converted[1])
df['lon'] = lon
df['lat'] = lat

# 제대로 된 위경도가 나오지 않음
# 서울특별시청 04524 서울특별시 중구 세종대로 110 전화번호 02-120 | 대표전화 02-2133-1271 | 빅데이터담당관(서소문청사)
# 전화필요
# 도로명 주소는 건물층수가 써져 있어 검색이 안됨
```

```python
# 사업자명으로 해안부 도로명주소 API 활용
import requests
from urllib.parse import quote
```

```python
with open('APIKey/roadapikey.txt') as file:
    road_key = file.read()
len(road_key)
```

```python
# 한글은 반드시 인코딩해서 보내야 함(한글은 주소창을 통과할 수 없음)
quote('롯데시네마 노원수락산역')
```

```python
road_url = 'https://www.juso.go.kr/addrlink/addrLinkApiJsonp.do'
params1 = f'confmKey={road_key}&currentPage=1&countPerPage=10'
params2 = f"keyword={quote('서초구청')}&resultType=json"
url = f'{road_url}?{params1}&{params2}'
```

```python
result = requests.get(url)
result.status_code          # client와 server가 서로 연결되면 200을 반환
```

```python
import json
res = json.loads(result.text[1:-1])
res
#  '개방주소정보 > API신청 > 팝업API 또는 검색API'
```
